### 2. 반복문
- 반복되는 코드를 실행할때 사용
- while, for, break, continue
- list comprehention

````python
# while
data = 3
while data: # 조건이 False 가 될때까지 구문의 코드를 실행
    # 반복되는 코드
    print(data)
    data -= 1
````

````python
# 학생이 국어:80점, 영어:90점, 수학: 100점 while문을 이용해서 총점과 평균을 출력
# 학생의 점수는 list, dict 표현
````

````python
score = {
    "Korean":80,
    "English":90,
    "Math":100,
}
````

````python
total, avg = 0,0
datas = list(score.values()).copy()
````

````python
while datas:
    total += datas.pop()


print("총점: "+str(total))
print("평균: "+str((total / len(score))))
````

````python
# 무한 루프
# break : 반복문을 중단 시킬때 사용되는 예약어
result = 1
while result:

    if result >= 10:
        break

    result += 1

print(result)
````

### for
- iterable한 값을 하나씩 꺼내서 value에 대입시킨후 코드를 iterable 변수의 값 갯수만큼 실행
```
for <variable> in <iterables>:
    <code>
```

````python
# for : countinue : 조건부분으로 올라가서 코드가 실행
ls = [0,1,2,3,4,]
for data in ls:

    if data % 2:    #data 가 홀수가 되면 continue 를 실행
        continue
                    # data가 짝수이면 print를 실행
    print(data, end=" ")
````

````python
# for를 이용하여 코드를 100번실행
# range 함수
list(range(100))
result = 0
for data in range(100):
    result += data
result
````

````python
# offset index 개념과 비슷하게 사용
list(range(5)), list(range(5,10)), list(range(10,1,-2))
````

````python
# 0~10 까지 짝수를 더한 총합
result = 0
for number in range(0, 11, 2):
    result += number
result
````

````python
score = {
    "Korean":80,
    "English":90,
    "Math":100,
}
````

````python
list(score.items())
````

````python
# for 문에서 iterabe한 데이터가 tuple 이면 여러개의 변수로 받을 수 있다.

for s, p in score.items():
    print(s,p)
````

- 구구단 출력(가로 출력)

    - print : end="\t"

### 3. List Comprehention
- 리스트 데이터를 만들어주는 방법
- for문 보다 빠르게 동작

````python
# 각각 값에 제곱한 결과 출력
ls = [0,1,2,3]
result = []
for data in ls:
    result.append(data**2)
result
````

````python
result = [data**2 for data in ls]
result
````

````python
ls = [0,1,2,3]
# result = ["짝수", "홀수", "짝수", "홀수"]

result = [
    "홀수" if data % 2 else "짝수"
    for data in ls
]
result
````

````python
# 리스트 컴프리핸션 조건문
ls = range(10)
[data for data in ls if data % 2]
````

````python
ls = [1,2,3]
dir(ls)
````

````python
[func for func in dir(ls) if func[:2] !="__" and func[0]=="c"]
````
